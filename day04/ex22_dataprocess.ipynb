#### 결측치(Missing Value)
- 데이터 수집과정에서 누락되거나 측정되지 않은 빈 값
- 데이터 분석에서는 결측치를 처리하는 일이 가장 중요

````python
# 새 컬럼 만들기
df['수학_1'] = df['수학'] - 5
````

````python
df
````

````python
# 정렬
df.sort_values('수학_1')
````

````python
df.sort_values('수학_1', ascending = False)
````

````python
# 결측치가 있는 데이터 삭제, 위험!
df1 = df.dropna()
df1
````

````python
# 결측치가 있는 값은 모두 같은 값으로 대체
df2 = df.fillna(45)
df2
````

````python
# 그룹화(groupby) - SQL의 Group By와 동일
df.groupby('수학').count()
````

````python
# 이름같은 str 타입은 통계치를 낼 수 없음
df[['국어', '영어', '수학']].fillna(45).groupby('수학').mean()
````

#### 간단 예제

````python
# 데이터프레임 생성
data = {
    'name' : ['민주', '철수', '영희', '길동'],
    'score' : [80, 95, 60, 70]
}

df = pd.DataFrame(data)
df
````

````python
# CSV로 쓰기
df.to_csv('./scores.csv', encoding = 'utf-8')
````

````python
# Excel로 쓰기
df.to_excel('./scores.xlsx')
````

````python
df['score'].mean()
````

````python
# 70점 이상만 추출
df['score'] >= 70
````

````python
df[df['score'] >= 70]
````

````python
# 합격 여부
# 리스트 컴프리헨션으로 새 컬럼 생성
df['result'] = ['PASS' if s >= 70 else 'FAIL' for s in df['score']]
````

````python
df
````
